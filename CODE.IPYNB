```python
# Normalize the feature values
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd

# Assuming 'df' is your original DataFrame
X = df[features].values

# Identify the non-numeric features
non_numeric_features = []
for i, col in enumerate(df[features].columns):
    if df[col].dtype == object:
        non_numeric_features.append(i)

# Encode the non-numeric features using one-hot encoding
X_encoded = X.copy()
for i in non_numeric_features:
    unique_values = df[features].iloc[:, i].unique()
    for value in unique_values:
        col_name = f"{features[i]}_{value}"
        X_encoded[:, i][X_encoded[:, i] == value] = 1
        X_encoded[:, i][X_encoded[:, i] != value] = 0
        df[col_name] = X_encoded[:, i]

# Remove the original non-numeric features
X = np.delete(X_encoded, non_numeric_features, axis=1)

# Normalize the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Normalize the feature values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Determining Optimal k')
plt.show()
```

```python
# Train the KNN clustering model
num_clusters = 4  # Based on the elbow plot
kmeans = KMeans(n_clusters=num_clusters, init='k-means++', random_state=42)
y_pred = kmeans.fit_predict(X_scaled)

# Assign cluster labels to the original dataframe
df['Cluster'] = y_pred
```

```python
#Evaluate the clustering model
num_clusters = 4  # Determined from the elbow plot
kmeans = KMeans(n_clusters=num_clusters, init='k-means++', random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score

# Calculate the silhouette scores
silhouette_avg = silhouette_score(X_scaled, labels)
sample_silhouette_values = silhouette_samples(X_scaled, labels)

# Plot the silhouette scores
plt.figure(figsize=(10, 8))
y_lower = 10
for i in range(num_clusters):
    ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    color = plt.cm.nipy_spectral(float(i) / num_clusters)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

plt.title("The silhouette plot for the various clusters.")
plt.xlabel("The silhouette coefficient values")
plt.ylabel("Cluster label")
plt.yticks([])
plt.xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
calinski_harabasz = calinski_harabasz_score(X_scaled, labels)
print(f"Calinski-Harabasz Score: {calinski_harabasz:.2f}")
```
